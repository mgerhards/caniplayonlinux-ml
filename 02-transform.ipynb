```python
import re

def extract_cpu_info(cpu_string):
    # Regular expressions for matching CPU information
    intel_pattern = re.compile(r'.*?(Intel).*?(Core|Xeon|Pentium|Celeron|Atom|i\d|\d{3,4})(?:.*?\s*@\s*[0-9\.]+GHz)?', re.IGNORECASE)
    amd_pattern = re.compile(r'.*?(AMD).*?(Ryzen|EPYC|Athlon|Athlon II|FX|Phenom|Opteron|A\d|Sempron|Athlon X\d|Athlon Silver|Athlon Gold|E\d|Custom APU|Phenom II|Threadripper|GX)(?:\s*-?\s*(\d{1,4}))?(?:.*?\s*@\s*[0-9\.]+GHz)?', re.IGNORECASE)
    pentium_pattern = re.compile(r'.*?(Intel).*?(Pentium Dual-Core)\s*(E\d{4}|T\d{4})', re.IGNORECASE)
    
    # Matching with Intel pattern
    intel_match = intel_pattern.match(cpu_string)
    if intel_match:
        manufacturer = intel_match.group(1)
        model = intel_match.group(2)
        generation_match = re.search(r'(\d{4})', cpu_string)
        if generation_match:
            generation = generation_match.group(1)[:2]  # Extract first two digits for 4-digit models
        else:
            generation = 'Unknown'
        return f"{manufacturer}, {model}, Generation: {generation}"
    
    # Matching with AMD pattern
    amd_match = amd_pattern.match(cpu_string)
    if amd_match:
        manufacturer = amd_match.group(1)
        model = amd_match.group(2)
        generation_match = amd_match.group(3)
        generation = generation_match[0] if generation_match else 'Unknown'  # Extract the first digit as generation for models with numbers
        return f"{manufacturer}, {model}, Generation: {generation}"
    
    # Matching with Pentium Dual-Core pattern
    pentium_match = pentium_pattern.match(cpu_string)
    if pentium_match:
        manufacturer = pentium_match.group(1)
        model = pentium_match.group(2)
        generation = 'Unknown'  # Generation is unknown for these models
        return f"{manufacturer}, {model}, Generation: {generation}"
    
    return None
```

```python
for cpu in cpu_values:
    cls = extract_cpu_info(cpu)
    if cls == None:
        print(f"{cpu} -> {cls}")
```
